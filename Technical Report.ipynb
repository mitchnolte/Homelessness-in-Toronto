# Description of Data
The Toronto Shelter System Flow data set is about homeless people in Toronto that use city-funded shelter services. It includes data about age, gender, and housing status, and splits the population into various groups. The population groups are as follows (definitions provided from the Open Data database, Toronto Shelter System Flow - City of Toronto Open Data Portal):
1. All populations: Refers to total number of individuals represented in the report without any demographic breakdown.
2. Chronic (refers to chronic homelessness): People who meet one of the two following criteria, as per the federal definition of chronic homelessness. The person has recorded a minimum of 180 overnight stays in the past year (365 days); or the person has recurrent overnight stays over the past three years with a cumulative duration of at least 546 nights.
3. Families: Individuals who are recorded as staying in a family designated overnight service.
4. Youth: Refers to unaccompanied youth and includes people who are between 16 and 24 years old by the last date of the reporting month and are not members of a family as defined above.
5. Single Adult: Refers to individuals who are neither classified as youth nor are members of a family.
6. Refugees: People who either identify as refugees upon intake to a shelter system or whose intake is completed into a program designated for refugees. Applies to all household members as answered by the household head.
7. Non-Refugees: People who are not categorized as refugees per the definition above.
8. Indigenous: An individual is classified as Indigenous if the person has self-identified as First Nations (status, non-status, treaty, non-treaty), Metis or Inuit in at least one of the intake(s) completed in SMIS (Shelter Management Information System).

### Data set Attributes
The following descriptions were provided from the Open Data database, Toronto Shelter System Flow - City of Toronto Open Data Portal:
1. date(mmm-yy): The time period (month/year) for which the data is being published (format: mmm-yy) 
2. population_group: Each row represents a different group: all population, chronic, refugees, families, youth, single adult and non-refugees 
3. returned_from_housing: Individuals who were previously recorded as “Moved to Permanent Housing” (see definition below) and returned to an overnight service in the reporting month. 
4. returned_to_shelter: People who were previously using the shelter system, then did not use the system for 3 months or longer, and have now returned. Some other communities may call this indicator “Returned from Inactive” 
5. newly_identified: People who entered the shelter system for the first time 
6. moved_to_housing: Moved to Permanent Housing: People who were using the shelter system and have moved to permanent housing. 
7. no_recent_shelter_use: People who were using the shelter system previously, but have not used the shelter system in the past three months. Some other communities may call this indicator “Moved to Inactive”. 1 
8. actively_homeless: People who have used the shelter system at least one time in the past three months and did not move to permanent housing. 
9. ageunder16: Age is calculated using the self-reported date of birth and the time the data was extracted form SMIS (Shelter Management Information System). AgeUnder16: less than 16 years of age
10. age16-24: Age is calculated using the self-reported date of birth and the time the data was extracted form SMIS (Shelter Management Information System). Age 16-24: between 16 and 24 years of age, inclusive
11. age25-44: Age is calculated using the self-reported date of birth and the time the data was extracted form SMIS (Shelter Management Information System). Age 25-44: between 25 and 44 years of age, inclusive
12. age45-64: Age is calculated using the self-reported date of birth and the time the data was extracted form SMIS (Shelter Management Information System). Age 45-64: between 45 and 64 years of age, inclusive
13. age65over: Age is calculated using the self-reported date of birth and the time the data was extracted form SMIS (Shelter Management Information System). AgeOver65: 65 years old and over
14. gender_male: Self-reported gender status during intake and as registered in SMIS. Identified as male gender during intake 
15. gender_female: Self-reported gender status during intake and as registered in SMIS. Identified as female during intake 
16. gender_transgender,non-binary_or_two_spirit: Self-reported gender status during intake and as registered in SMIS. Identified as transgender, non-binary or two-spirit during intake 
17. population_group_percentage: Each row represents the percentage of each population group (chronic, refugees, families, youth, single adult and non-refugees). The proportion is calculated from all the population. 

		
The data is updated monthly with a new entry for each population group. The data was collected from the Shelter Management Information System (SMIS), which is used by Toronto’s shelter services to take people into their programs.
```python
population = list(filter(lambda x: x['population_group'] == 'All Population', data))
age_groups = []
age_groups.append({'age': "Under 16", "amount": reduce(lambda total,x: total+int(x['ageunder16']), population, 0)})
age_groups.append({'age': "16-24", "amount": reduce(lambda total,x: total+int(x['age16-24']), population, 0)})
age_groups.append({'age': "25-44", "amount": reduce(lambda total,x: total+int(x['age25-44']), population, 0)})
age_groups.append({'age': "45-64", "amount": reduce(lambda total,x: total+int(x['age45-64']), population, 0)})
age_groups.append({'age': "65+", "amount": reduce(lambda total,x: total+int(x['age65over']), population, 0)})

df2 = pd.DataFrame(age_groups)
df2.plot(x='age', y='amount', xlabel="Age Group", ylabel="Total Num. Actively Homeless", title="Active Homelessness per Age Group", figsize=(15,8))
plt.show()

age_max = age_groups[0]
for i in range(1, len(age_groups)):
    if age_groups[i]['amount'] > age_max['amount']:
        age_max = age_groups[i]
print("\nThe age group most susceptible to homelessness is", age_max['age']+".")
```

### 2. What was the average of active homelessness for people under 16 years of age in the year 2020? Has this improved in 2021?

```python
population = filter(lambda x: x['population_group'] == 'All Population', data)
pop2020 = list(filter(lambda x: x['date(mmm-yy)'][-2:] == '20', population))

print("2020:\nJan\tFeb\tMar\tApr\tMay\tJun\tJul\tAug\tSep\tOct\tNov\tDec")
for x in pop2020:
    print(x['ageunder16'], end="\t")
    
avg20 = round(np.mean(list(map(lambda x: int(x['ageunder16']), pop2020))))

population = filter(lambda x: x['population_group'] == 'All Population', data)
pop2021 = list(filter(lambda x: x['date(mmm-yy)'][-2:] == '21', population))

print("\n\n2021:\nJan\tFeb\tMar\tApr\tMay\tJun\tJul\tAug\tSep")
for x in pop2021:
    print(x['ageunder16'], end="\t")

avg21 = round(np.mean(list(map(lambda x: int(x['ageunder16']), pop2021))))

print("\n\nThe average number of actively homeless people under 16 years of age in 2020 was %d." % avg20)
print("This number ", end="")
if avg21 < avg20:
    print("improved in 2021, ", end="")
else:
    print("has gotten worse in 2021, ", end="")
print("with a new average of %d." % avg21)
```

### 3. Which months of the year is the population most likely to return to shelter? List the top 3 months.

```python
months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug" ,"Sep", "Oct","Nov","Dec"]
def monthlyAvg(x):
    averages = {k:[] for k in months}
    for i in x:
        if i['population_group'] == 'All Population':
            total = int(i['returned_from_housing']) + int(i['returned_to_shelter']) 
            averages[i['date(mmm-yy)'][:3]].append(total)
    for k,v in averages.items():
        averages[k] = sum(v) / len(v)
    return averages

averages = monthlyAvg(data)
yAxis = list(averages.values())
xAxis = months
plt.bar(xAxis, yAxis, color = 'teal')
plt.xlabel("Months")
plt.ylabel("Population")
plt.title("Monthly Average Returned To Shelter Population")
plt.show
averages = sorted(averages.items(), key = lambda x: x[1], reverse = True)
print("July, December and August have the highest amount of people returning to shelter with the following numbers;")
print(averages[:3])
```

Our third question had us searching for which months had the most population returning to the shelter. Iterating through the data set we stored the total population in each month as a value in a dictionary with months as the key. If there were multiple years worth of data, there would be multiple values for the key and so we just took the average of those values per key. Only one line of code was changed to create the upcoming two graphs(line 6):

Using the above line of code we generated the first graph to sum the total amount of homeless people per month.

### 4. Find the ratio of people that have moved into housing per population group, do certain groups have greater success than others?

```python
import pandas as pd
import numpy as np

df = pd.read_csv("toronto-shelter-system-flow.csv")

#Get the total number of people per population group (this will get ratio of people moving into housing per population group)
df["total_group_population"] = df["gender_male"] + df["gender_female"] + df["gender_transgender,non-binary_or_two_spirit"]

#Get ratio of people moving into housing per total group population
df ["Ratio people moved into housing per population group"] = df["moved_to_housing"]/df["total_group_population"]

#Pivot table to find the mean of ratios
df_ratio_mean_pivot = pd.pivot_table(df, values ='Ratio people moved into housing per population group', index='population_group', aggfunc ='mean', fill_value = 0)

#Sort in descending order for easier analysis
df_ratio_mean_pivot = df_ratio_mean_pivot.sort_values(by=['Ratio people moved into housing per population group'])

#Remove irrelevant column
df_ratio_mean_pivot.drop('All Population')
```

Question four calculates the ratio of people moving into housing per population group. The solution was achieved by the following technical steps:
1. Import Pandas, which is the only library needed
2. Create a dataframe from the data set using pandas.read_csv (data set)
3. Create a new column in the data set named total_group_population. The values in this column will be the addition of males in row, females in row, transgenders in row, non-binary in row, and two spirit in row; this would constitute all possible people in the data set per row (per row is equivalent to saying per population group in this data set)
4. Create a new column named “Ratio people moved into housing per population group”. The values in this column will be the division of people moving into housing per row by the total population group (step C) per row; this will give the success ratio of a specific population group moving into housing
5. Use Pandas Pivot Table function to calculate the mean (solution) of all the outputs per population group
6. Sort the means in order for readability and analysis

The results of question four are listed below, population groups ordered from highest to lowest in successfully moving in housing.
1. Families (~ 13.0% success rate)
2. Refugees (~ 11.5% success rate)
3. Chronic (~ 6.5% success rate)
4. Youth (~ 5.0% success rate)
5. Non-refugees (~ 3.5% success rate)
6. Single Adult (~ 2.5% success rate)
